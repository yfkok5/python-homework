```python
# -*- coding: UTF-8 -*-
"""PyRamen Homework Starter."""

# @TODO: Import libraries
import csv
from pathlib import Path

# @TODO: Set file paths for menu_data.csv and sales_data.csv
menu_filepath = Path(r'C:\Users\yfkok\OneDrive\Desktop\Monash Bootcamp\python-homework\PyRamen\Resources\menu_data.csv')
sales_filepath = Path(r'C:\Users\yfkok\OneDrive\Desktop\Monash Bootcamp\python-homework\PyRamen\Resources\sales_data.csv')

filename = open(menu_filepath, 'r')
 
# creating dictreader object
menu = csv.DictReader(filename)
 
# creating empty lists
item = []
category = []
description = []
price = []
cost = []
all_elements = ['item','category','description','price','cost']
# iterating over each row and append
# values to empty list

for col in menu:
    item.append(col['item'])
    category.append(col['category'])
    description.append(col['description'])
    price.append(float(col['price']))
    cost.append(int(col['cost']))

#zipping values together    
    value = zip(category,description,price,cost
#assigning item as the key to menu_dict
    menu_dict = dict(zip(item, value))   
```

```python
filename = open(sales_filepath, 'r')
 
# creating dictreader object
sales = csv.DictReader(filename)
 
# creating empty lists
menu_item = []
quantity = []

# iterating over each row and append
# values to empty list
for col in sales:
    
    menu_item.append(col['Menu_Item'])
    quantity.append(int(col['Quantity']))

sales_list= list(zip(menu_item, quantity))
```

```python
quantity_spicymiso = []
quantity_misocrabramen = []
quantity_nagomishoyu = []
quantity_shioramen = []
quantity_softshellcrab = []
quantity_burntgarlic = []
quantity_tonkotsu = []
quantity_toripaitan = []
quantity_trufflebutter = []
quantity_vegcurrykingtrumpet = []
quantity_vegspicymiso = []

sum_quantity_food = []
#Iterate through sales list to obtain quantity of food
for n in range(len(sales_list)):
   
    if sales_list[n][0] == 'spicy miso ramen':
        quantity_spicymiso.append(sales_list[n][1])
    elif sales_list[n][0] == 'miso crab ramen':
        quantity_misocrabramen.append(sales_list[n][1])
    elif sales_list[n][0] == 'burnt garlic tonkotsu ramen':
        quantity_burntgarlic.append(sales_list[n][1])
    elif sales_list[n][0] == 'nagomi shoyu':
        quantity_nagomishoyu.append(sales_list[n][1])
    elif sales_list[n][0] == 'shio ramen':
        quantity_shioramen.append(sales_list[n][1])
    elif sales_list[n][0] == 'soft-shell miso crab ramen':
        quantity_softshellcrab.append(sales_list[n][1])
    elif sales_list[n][0] == 'tonkotsu ramen':
        quantity_tonkotsu.append(sales_list[n][1])
    elif sales_list[n][0] == 'tori paitan ramen':
        quantity_toripaitan.append(sales_list[n][1])
    elif sales_list[n][0] == 'truffle butter ramen':
        quantity_trufflebutter.append(sales_list[n][1])                                                
    elif sales_list[n][0] == 'vegetarian curry + king trumpet mushroom ramen':
        quantity_vegcurrykingtrumpet.append(sales_list[n][1])                                                
    elif sales_list[n][0] == 'vegetarian spicy miso':
        quantity_vegspicymiso.append(sales_list[n][1])                                                

sum_quantity_food = []

sum_quantity_food.append(sum(quantity_spicymiso))
sum_quantity_food.append(sum(quantity_misocrabramen))
sum_quantity_food.append(sum(quantity_burntgarlic))
sum_quantity_food.append(sum(quantity_nagomishoyu))
sum_quantity_food.append(sum(quantity_shioramen))
sum_quantity_food.append(sum(quantity_softshellcrab))
sum_quantity_food.append(sum(quantity_tonkotsu))
sum_quantity_food.append(sum(quantity_toripaitan))
sum_quantity_food.append(sum(quantity_trufflebutter))
sum_quantity_food.append(sum(quantity_vegcurrykingtrumpet))
sum_quantity_food.append(sum(quantity_vegspicymiso))

print(sum_quantity_food)
                                          
```

```python
#not printing it as it contains too much data
price_food=[]
cogs_food=[]
ramen = ('spicy miso ramen',
         'miso crab ramen',  
         'burnt garlic tonkotsu ramen',
         'nagomi shoyu',
         'shio ramen',
         'soft-shell miso crab ramen',
         'tonkotsu ramen',
         'tori paitan ramen',
         'truffle butter ramen',
         'vegetarian curry + king trumpet mushroom ramen',                                              
         'vegetarian spicy miso',)  

def pricefood(k):
    price_food.append(menu_dict[k][2])

def cogsfood(k):
    cogs_food.append(menu_dict[k][3])

for n in ramen:
    pricefood(n)
    cogsfood(n)   

print(price_food)
print(cogs_food)
```

```python
compiled = zip(sum_quantity_food,price_food,cogs_food)
compiled_dict = dict(zip(ramen, compiled)) 

revenue = []
cogs = []
profit = []

#formulas to determine revenue, cogs and profits
for n in ramen:
    revenue_new = compiled_dict[n][0]*compiled_dict[n][1]
    cogs_new = compiled_dict[n][0]*compiled_dict[n][2]
    profit_new = revenue_new-cogs_new
    
    cogs.append(cogs_new)
    revenue.append(revenue_new)
    profit.append(profit_new)

#compiled revenue + cogs + profits as "values"  
results_compiled = zip(sum_quantity_food,revenue,cogs,profit)

#creating a dictionary for the final results by putting ramen as keys
results_dict = dict(zip(ramen,results_compiled))

for n in ramen:
    print(f'{n} "01-count": {results_dict[n][0]},"02-revenue": {results_dict[n][1]},"03-cogs": {results_dict[n][2]},"04-profit": {results_dict[n][3]}')
```
